# Step 1: Initialize Population

```python
import random
def initPopulation(b, n ): # n is the number of solutions, b is the length, 1st papameter num of bitd 
    p = [random.choices([0,1],k = b) for i in range(n)]  # k-b tah row repet r jono use kora hoise &list compihantion &
    return p
```

```python
initPopulation(5,6) # aktah list r modde ar aktah list
```

# Step 2: Reproduction or Parent Selection

Fitness Calculation

```python


def binToDecimal(binList):   # bin to des convert its a functon 
  decValue = 0 # decimel value
  power = len(binList)
  for digit in binList: # binary to decimel
    decValue += digit*2**(power-1)
    power -= 1
  return decValue

def func(x):
  return x**2 - 11*x + 150

def getFitness(p): # p is the whole population
    f = [] # to store fitness values
    for sol in p:
        d = binToDecimal(sol)
        v = func(d)
        if v< 0: f.append(0)
        else: f.append(v) # fitness function
    return f
```

```python
population = initPopulation(5,6)
getFitness(population)
```

calculate probability and select parents

```python
import numpy as np
def select_parent(n, pop): # probability, parent selection koitah parent ami nibo tah select 
    # probability
    fitness = getFitness(pop)
    total_fitness = sum(fitness)
    prob = [f/total_fitness for f in fitness]
    cumSumProb = np.cumsum(prob)   # cumalatic sum bin r last value ber korar jonno use kora hoise
    #print(cumSumProb)
    # parent selection
    parents = []
    for i in range(n):
        # roullete wheel, generate a random number
        r = random.random()
        #print(r)
        # check bin number of r        
        for j in range(len(cumSumProb)):
            if r <= cumSumProb[j] :#checking the bin
                #print(j)
                parents.append(pop[j]) # jaitah soto oitah nisse means match hosse oitah nibo
                break
    return parents
```

```python
population = initPopulation(5,6)
print(population)
getFitness(population)
parents = select_parent(4, population)
print(parents)
```

# Step 3: Crossover

```python
def crossover(parents): # define a method crossover (input is parents set)

    #num_cross = len(parents)/2
    #print(num_cross)
    offsprings = []

    for c in range(0, len(parents), 2): # for loop to perform crossovers
        p = parents[c]
        q = parents[c+1]
        
        cp = random.choice(range(1,7))
        print(cp)
        # perform crossover
        offsprings.append(p[0:cp] + q[cp:len(p)])
        offsprings.append(q[0:cp] + p[cp:len(p)])
        
    return offsprings
    # return the offsprings


```

# Step 5: Select Survivor
The selected survivors (solutions) will be added to population

Select top 2 offspring based on fitness value and add them to population.

```python
# define a method to select top offsprings (input offsprings, top)
def top_offspring(offsprings, top):
    top_offsprings = []
    survivors = sortSurvivors = getFitness(offsprings)
    sortSurvivors.sort(reverse=True)
    for t in range(top):
        temp = offsprings[survivors.index(sortSurvivors[t])]
        #temp = offsprings[survivors.index(max(survivors))]
        print(temp)
        population.append(temp)
    
top_offspring(offsprings, 2)
```
